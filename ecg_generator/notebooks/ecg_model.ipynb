```python
import numpy as np
```

```python
def gaussian(t, mu, sigma, amplitude):
    """
    Gaussian wave function for ECG components
    """
    return amplitude * np.exp(-((t - mu)**2) / (2 * sigma**2))
```

```python
def ecg_beat(t):
    """
    Single ECG heartbeat as sum of Gaussian waves
    """
    P = gaussian(t, 0.2, 0.025, 0.2)
    Q = gaussian(t, 0.35, 0.010, -0.3)
    R = gaussian(t, 0.40, 0.012, 1.2)
    S = gaussian(t, 0.45, 0.010, -0.4)
    T = gaussian(t, 0.65, 0.04, 0.35)
    return P + Q + R + S + T
```
