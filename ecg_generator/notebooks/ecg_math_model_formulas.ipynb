# Mathematical Model of ECG

This notebook shows the **mathematical formulation** of an ECG beat using Gaussian functions.

The ECG waveform can be modeled as a sum of Gaussian functions representing **P, Q, R, S, T waves**:

\\[
ECG(t) = P(t) + Q(t) + R(t) + S(t) + T(t)
\\]

where each wave is defined as a Gaussian:

\\[
G(t; \mu, \sigma, A) = A \cdot e^{-\\frac{(t - \mu)^2}{2 \sigma^2}}
\\]

- \\( \mu \\) = mean (center of the wave)  
- \\( \sigma \\) = standard deviation (width of the wave)  
- \\( A \\) = amplitude of the wave

## P Wave

\\[
P(t) = 0.2 \cdot e^{-\\frac{(t - 0.2)^2}{2 \cdot 0.025^2}}
\\]

Represents **atrial depolarization**.


## QRS Complex

- Q wave: \\( Q(t) = -0.3 \cdot e^{-\\frac{(t - 0.35)^2}{2 \cdot 0.01^2}} \\)  
- R wave: \\( R(t) = 1.2 \cdot e^{-\\frac{(t - 0.40)^2}{2 \cdot 0.012^2}} \\)  
- S wave: \\( S(t) = -0.4 \cdot e^{-\\frac{(t - 0.45)^2}{2 \cdot 0.01^2}} \\)  

Represents **ventricular depolarization**.

## T Wave

\\[
T(t) = 0.35 \cdot e^{-\\frac{(t - 0.65)^2}{2 \cdot 0.04^2}}
\\]

Represents **ventricular repolarization**.



## Complete Normal ECG Beat

Combining all waves for a **normal heartbeat**:

\\[
ECG_{\text{normal}}(t) = P(t) + Q(t) + R(t) + S(t) + T(t)
\\]

or explicitly:

\\[
ECG_{\text{normal}}(t) = 0.2 e^{-\\frac{(t-0.2)^2}{2 \cdot 0.025^2}} 
- 0.3 e^{-\\frac{(t-0.35)^2}{2 \cdot 0.01^2}} 
+ 1.2 e^{-\\frac{(t-0.40)^2}{2 \cdot 0.012^2}} 
- 0.4 e^{-\\frac{(t-0.45)^2}{2 \cdot 0.01^2}} 
+ 0.35 e^{-\\frac{(t-0.65)^2}{2 \cdot 0.04^2}}
\\]

This represents a **regular ECG beat**.



## Arrhythmic ECG Beat

In arrhythmia, the **timing (mean \\( \mu \\))** and **amplitude (A)** of each wave can vary:

\\[
ECG_{\text{arrhythmia}}(t) = P(t - \delta_P) + Q(t - \delta_Q) + R(t - \delta_R) + S(t - \delta_S) + T(t - \delta_T)
\\]

or explicitly:

\\[
ECG_{\text{arrhythmia}}(t) = 0.2' e^{-\\frac{(t-0.2 - \delta_P)^2}{2 \cdot 0.025^2}} 
- 0.3' e^{-\\frac{(t-0.35 - \delta_Q)^2}{2 \cdot 0.01^2}} 
+ 1.2' e^{-\\frac{(t-0.40 - \delta_R)^2}{2 \cdot 0.012^2}} 
- 0.4' e^{-\\frac{(t-0.45 - \delta_S)^2}{2 \cdot 0.01^2}} 
+ 0.35' e^{-\\frac{(t-0.65 - \delta_T)^2}{2 \cdot 0.04^2}}
\\]

- \\( \delta_P, \delta_Q, \delta_R, \delta_S, \delta_T \\) are **small random shifts in timing**.  
- Amplitudes may also vary: \\( 0.2', 1.2', -0.4', 0.35' \\).  

This models **irregular heartbeat intervals**.


### Area under R wave (Normal vs Arrhythmia)

- Normal R-wave:

\\[
\int_{-\infty}^{\infty} R(t) \, dt = 1.2 \cdot \sigma \sqrt{2 \pi} \approx 0.144
\\]

- Arrhythmic R-wave (amplitude = 1.0, width = 0.014):

\\[
\int_{-\infty}^{\infty} R_{\text{arr}}(t) \, dt = 1.0 \cdot 0.014 \sqrt{2 \pi} \approx 0.035
\\]

This shows how **arrhythmia changes both timing and energy of the ECG waves**.
